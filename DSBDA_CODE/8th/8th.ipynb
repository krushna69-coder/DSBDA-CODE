```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load the Titanic dataset from the CSV file
titanic_data = pd.read_csv('titanic.csv')

# Check the first few rows to ensure the data is loaded correctly
print(titanic_data.head())

# Plotting the distribution of 'fare' column using a histogram
plt.figure(figsize=(10, 6))
sns.histplot(titanic_data['Fare'], kde=True, color='skyblue', bins=30)
plt.title('Distribution of Ticket Prices (Fare) on Titanic')
plt.xlabel('Fare')
plt.ylabel('Frequency')
plt.show()
```
